```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("notebook\\data\\Breast_Cancer.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
X = df.iloc[:,:9]
Y = df["Status"]
```

```python
LE = LabelEncoder()
```

```python
X.Race = LE.fit_transform(X.Race)
X.Marital_Status = LE.fit_transform(X.Marital_Status)
X.tstage= LE.fit_transform(X.tstage)
X.N_Stage = LE.fit_transform(X.N_Stage)
X.Sixth_Stage = LE.fit_transform(X.Sixth_Stage)
X.differentiate = LE.fit_transform(X.differentiate)
X.A_Stage = LE.fit_transform(X.A_Stage)
X.Estrogen_Status = LE.fit_transform(X.Estrogen_Status)
X.Progesterone_Status = LE.fit_transform(X.Progesterone_Status)
```

```python
X
```

# Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=20)
```

```python
GB = GaussianNB()
GB.fit(x_train,y_train)
```

```python
GB.predict(x_test)
```

```python
GB.score(x_test,y_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
lor.fit(x_train,y_train)
```

```python
lor.predict(x_test)
```

```python
lor.score(x_test,y_test)
```

# K Nearest Neighbour 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors= 3)
knn.fit(x_train,y_train)
```

```python
knn.predict(x_test)
```

```python
knn.score(x_test,y_test)
```
